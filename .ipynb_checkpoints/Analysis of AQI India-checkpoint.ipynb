## Data Dictionary: Key Pollutants

| Column Name | Full Form | Description | Health Impact |
|-------------|------------|-------------|----------------|
| **PM2.5** | Particulate Matter ≤ 2.5 microns | Fine airborne particles from combustion, dust, and smoke. Can penetrate deep into lungs. | Linked to heart disease, stroke, and respiratory issues |
| **PM10** | Particulate Matter ≤ 10 microns | Larger particles like dust, pollen, and mold. Suspended in air. | Causes respiratory irritation and worsens asthma |
| **NO₂** | Nitrogen Dioxide | Emitted from vehicles and industrial processes. Contributes to smog. | Irritates lungs and increases risk of infections |
| **SO₂** | Sulfur Dioxide | Produced by burning coal and oil. Can form acid rain. | Triggers asthma and other respiratory problems |
| **CO** | Carbon Monoxide | Colorless gas from incomplete combustion. Highly toxic in enclosed spaces. | Reduces oxygen delivery in body; can be fatal at high levels |
| **O₃** | Ozone (Ground-level) | Formed by sunlight reacting with pollutants. Not emitted directly. | Causes chest pain, coughing, and worsens lung diseases |

### Experimenting on the Imputation Methods

```python
def impute_and_insert(df, features, group_keys, methods):
    for feature in features:
        for group_key in group_keys:
            for method in methods:
                # Build column name
                col_name = f"{feature}_imputed_{method.upper()}_by{group_key.capitalize()}"
                
                # Compute imputed values
                imputed_series = df[feature].fillna(
                    df.groupby(group_key)[feature].transform(method)
                )
                
                # Insert after original column
                insert_pos = df.columns.get_loc(feature) + 1
                df.insert(insert_pos, col_name, imputed_series)
```

```python
features = ['so2', 'no2', 'rspm']
group_keys = ['location', 'state', 'Season']
methods = ['mean', 'median']

impute_and_insert(temp_df, features, group_keys, methods)
```

### How are we comparing the scores?

**`Score = Total Deviation from Original`**

For each imputed column (e.g., so2_imputed_MEDIAN_bySeason), we compare it to the original column (so2) using three metrics:

1. Mean Difference
`mean_diff = abs(imputed_mean - original_mean)`

This tells us how much the central tendency has shifted.

2. Standard Deviation Difference
`std_diff = abs(imputed_std - original_std)`

This shows whether the spread of values has changed — important for preserving variability.

3. Median Difference
`median_diff = abs(imputed_median - original_median)`

This checks if the core of the distribution (50th percentile) is still aligned.

Final Score
`score = mean_diff + std_diff + median_diff`

**`Lower score = better preservation of original distribution.`**

Why This Works
- It’s interpretable, you can see exactly how much each method deviates.


```python
def compare_imputation_strategies(summary_df):
    # Identify original features
    originals = ['so2', 'no2', 'rspm']
    strategies = ['MEDIAN_bySeason', 'MEAN_bySeason',
                  'MEDIAN_byState', 'MEAN_byState',
                  'MEDIAN_byLocation', 'MEAN_byLocation']
    results = []
    for feature in originals:
        original_stats = summary_df.loc[feature]
        for strategy in strategies:
            col_name = f"{feature}_imputed_{strategy}"
            if col_name not in summary_df.index:
                continue
            imputed_stats = summary_df.loc[col_name]
            # Calculate absolute differences
            mean_diff = abs(imputed_stats['mean'] - original_stats['mean'])
            std_diff = abs(imputed_stats['std'] - original_stats['std'])
            median_diff = abs(imputed_stats['50%'] - original_stats['50%'])
            results.append({
                'Feature': feature,
                'Strategy': strategy,
                'Mean_Diff': round(mean_diff, 3),
                'Std_Diff': round(std_diff, 3),
                'Median_Diff': round(median_diff, 3)
            })
    # Convert to DataFrame
    result_df = pd.DataFrame(results)
    # Rank by closeness to original (lower is better)
    result_df['Score'] = result_df[['Mean_Diff', 'Std_Diff', 'Median_Diff']].sum(axis=1)
    result_df = result_df.sort_values(by=['Feature', 'Score'], ascending = True)

    return result_df
```

```python
comparison = compare_imputation_strategies(temp_df.drop(columns = ['date']).describe().T)
print(comparison)
```

We always choose the strategy with the least score, because it means the imputed values are closest to the original distribution in terms of mean, standard deviation, and median. **Lower = better fit.**

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=comparison.sort_values(by = 'Score',ascending = True), x='Strategy', y='Score', hue='Feature')
plt.xticks(rotation=45)
plt.title("Imputation Strategy Score Comparison")
plt.ylabel("Total Deviation from Original")
plt.legend(title='Feature')
plt.tight_layout()
plt.show()
```

It can be observed that **Season-wise median imputation** best preserves original mean, median, and spread across SO₂, NO₂, and RSPM. It minimizes distortion, handles outliers well and aligns with seasonal pollution patterns.

### Pollution Index

Excluded PM2.5 and SPM due to high missingness (97.86% and 54.48% respectively). The index is based on pollutants with reliable coverage: RSPM (PM10), NO₂ and SO₂.

**Pollution_Index = (0.45 × RSPM) + (0.35 × NO₂) + (0.20 × SO₂)**

Where:
- **RSPM** = Respirable Suspended Particulate Matter (PM10)
- **NO₂** = Nitrogen Dioxide
- **SO₂** = Sulphur Dioxide

Weights are derived from WHO and CPCB standards, reflecting relative health impact and data availability.

```python
weights_main = {
    'Respirable_Suspended_Particulate_Matter': 0.45,
    'Nitrogen_Dioxide': 0.35,
    'Sulphur_Dioxide': 0.20
}

data['Pollution_Index'] = (
    data['Respirable_Suspended_Particulate_Matter'] * weights_main['Respirable_Suspended_Particulate_Matter'] +
    data['Nitrogen_Dioxide'] * weights_main['Nitrogen_Dioxide'] +
    data['Sulphur_Dioxide'] * weights_main['Sulphur_Dioxide']
)
```

```python
def plot_n_polluted(subset, type = "most"):
    if type == "most":
        type_ = 'Reds'
    elif type == "least":
        type_ = 'Greens'
    else:
        type_ = 'winter'
    x_labels = subset.index.astype(str)
    y_values = subset['Pollution_Index'].values
    
    fig = go.Figure(go.Bar(
        x=x_labels,
        y=y_values,
        text=[f"{val:.2f}" for val in y_values],
        textposition='outside',
        marker=dict(color=y_values, colorscale = type_),
        hovertemplate="City: %{x}<br>Pollution Index: %{y:.2f}<extra></extra>"
    ))
    
    fig.update_layout(
        title = f"Pollution Index by City - Top {n} {type} Polluted Cities",
        xaxis_title="City",
        yaxis_title="Pollution Index",
        yaxis=dict(tickformat=".2f"),
        template="plotly_white",
        width=900,
        height=500
    )
    
    fig.show()
```

```python
subset_1 = result.iloc[0:n]
plot_n_polluted(subset_1, "most")

subset_2 = result.iloc[len(result) - (n+1):-1]
plot_n_polluted(subset_2[::-1], "least")
```

```python
def reporting_coverage_summary(df, city):
    """
    Calculates reporting coverage for each station in a city.
    Returns a DataFrame with first report date, total days reported/unreported, and percentages.
    """
    print(f"\nReport for the City : {B}{city}{R}.")
    # Filter for city and ensure Date is datetime
    city_df = df[df['City'] == city].copy()
    city_df['Date'] = pd.to_datetime(city_df['Date'])

    # Get last available date in dataset
    last_date = city_df['Date'].max()
    print(f"Last available date in dataset: {last_date.date()}\n\n")

    # Group by station
    summary = []
    for station, group in city_df.groupby('Location_Monitoring_Station'):
        first_date = group['Date'].min()
        total_days = (last_date - first_date).days + 1
        reported_days = group['Date'].nunique()
        unreported_days = total_days - reported_days
        pct_reported = round((reported_days / total_days) * 100, 2)
        pct_unreported = round((unreported_days / total_days) * 100, 2)

        summary.append({
            'Location_Monitoring_Station': station,
            'First_Report_Date': first_date.date(),
            'Total_Days': total_days,
            'Reported_Days': reported_days,
            'Unreported_Days': unreported_days,
            '% Reported': pct_reported,
            '% Unreported': pct_unreported
        })

    return pd.DataFrame(summary).sort_values('First_Report_Date', ascending = True).reset_index(drop = True)
```

|                        | 1 hour                 | 10.0 mg/m³          | 4.0 mg/m³                            | 2.0 mg/m³           |

---

**\*** Annual arithmetic mean of minimum 104 measurements in a year taken twice a week 24 hourly at uniform interval.  
**\*\*** 24 hourly / 8 hourly values should be met 98% of the time in a year. However, 2% of the time, it may exceed but not on two consecutive days.

**Source**: [Assam Pollution Control Board – NAAQS](https://pcb.assam.gov.in/information-services/national-ambient-air-quality-standards)

```python
def check_daily_pollutant_exceedance(data):
    import pandas as pd

    # Mapping area categories
    area_mapping = {
        'Residential, Rural and other Areas': 'Residential',
        'Residential and others'            : 'Residential',
        'Residential'                       : 'Residential',
        'Industrial Area'                   : 'Industrial',
        'Industrial Areas'                  : 'Industrial',
        'Industrial': 'Industrial',
        'Sensitive Area': 'Sensitive',
        'Sensitive Areas': 'Sensitive',
        'Sensitive': 'Sensitive'
    }
    data['Area Type'] = data['Area_Category'].map(area_mapping)

    # Safe limits
    limits = {
        'Industrial': {'SO2': 120, 'NO2': 120, 'RSPM': 150},
        'Residential': {'SO2': 80, 'NO2': 80, 'RSPM': 100},
        'Sensitive': {'SO2': 30, 'NO2': 30, 'RSPM': 75}
    }

    # Group and average by City, Date, AreaType
    avg_df = data.groupby(['City', 'Date', 'Area Type'], as_index=False)[
        ['Sulphur_Dioxide', 'Nitrogen_Dioxide', 'Respirable_Suspended_Particulate_Matter']
    ].mean()

    # Check exceedance and collect results
    def check_exceed(row, pollutant, colname):
        limit = limits[row['Area Type']][pollutant]
        value = row[colname]
        if value > limit:
            return {
                'City': row['City'],
                'Date': row['Date'],
                'Area Type': row['Area Type'],
                pollutant: round(value, 2),
                '% Above Limit': round(((value - limit) / limit) * 100, 2)
            }

    def filter_consecutive_exceedance(records, pollutant):
        df = pd.DataFrame(records)
        if df.empty:
            return df
        df['Date'] = pd.to_datetime(df['Date'])
        df = df.sort_values(['City', 'Date'])
        result = []

        for city, group in df.groupby('City'):
            group = group.reset_index(drop=True)
            group['Gap'] = group['Date'].diff().dt.days.fillna(0).astype(int)
            group['Block'] = (group['Gap'] != 1).cumsum()
            block_sizes = group.groupby('Block').size()
            group['Consecutive_Days'] = group['Block'].map(block_sizes)
            result.append(group[group['Consecutive_Days'] >= 2])

        df_final = pd.concat(result).drop(columns=['Gap', 'Block'])
        cols = ['City', 'Date', 'Area Type', pollutant, '% Above Limit', 'Consecutive_Days']
        return df_final[cols]

    so2_exceed, no2_exceed, rspm_exceed = [], [], []
    for _, row in avg_df.iterrows():
        if res := check_exceed(row, 'SO2', 'Sulphur_Dioxide'):
            so2_exceed.append(res)
        if res := check_exceed(row, 'NO2', 'Nitrogen_Dioxide'):
            no2_exceed.append(res)
        if res := check_exceed(row, 'RSPM', 'Respirable_Suspended_Particulate_Matter'):
            rspm_exceed.append(res)
            
    print(f"\nStandard limits used for SO2 - \n\tIndustrial Area   : 120 \tResidential Area  : 80  \tSensitive Area    : 30")
    so2_df = filter_consecutive_exceedance(so2_exceed, 'SO2')
    so2_df = so2_df.sort_values(by='% Above Limit', ascending=False).reset_index(drop = True)
    print(f"{B}{len(so2_df)}{R} records found with SO2 exceeding limits on 2+ consecutive days.")

    print(f"\n\nStandard limits used for NO2 - \n\tIndustrial Area   : 120 \tResidential Area  : 80  \tSensitive Area    : 30")
    no2_df = filter_consecutive_exceedance(no2_exceed, 'NO2')
    no2_df = no2_df.sort_values(by='% Above Limit', ascending=False).reset_index(drop = True)
    print(f"{B}{len(no2_df)}{R} records found with NO2 exceeding limits on 2+ consecutive days.")

    print(f"\n\nStandard limits used for RSPM - \n\tIndustrial Area   : 150 \tResidential Area  : 100 \tSensitive Area    : 75")
    rspm_df = filter_consecutive_exceedance(rspm_exceed, 'RSPM')
    rspm_df = rspm_df.sort_values(by='% Above Limit', ascending=False).reset_index(drop = True)
    print(f"{B}{len(rspm_df)}{R} records found with RSPM exceeding limits on 2+ consecutive days.")
    return so2_df, no2_df, rspm_df
```

## Which states have the most polluted cities?

Wee need some clarification for answering the question.
 - **`most polluted`** : Is it based on average pollution levels, frequency of exceedance, or number of cities breaching limits?
 - Specify pollutants of interest: `SO₂`, `NO₂`, `RSPM`, `PM2.5` or a composite index like `Pollution_Index`
 - Determine the time frame: `Daily`, `monthly`, `yearly`? `Seasonal` variation?

---
Since we are given a simple problem statement with `most polluted cities`, we will rank the states based on `Pollution_Index` of **top `n`** polluted cities.

```python
def rank_states_by_yearly_pollution(data):
    daily_city_pollution = (
        data.groupby(['City', 'State', 'Date'])['Pollution_Index']
        .mean()
        .reset_index()
    )
    daily_city_pollution['Year'] = daily_city_pollution['Date'].dt.year
    yearly_city_pollution = (
        daily_city_pollution.groupby(['City', 'State', 'Year'])['Pollution_Index']
        .mean()
        .reset_index()
    )
    state_yearly_pollution = (
        yearly_city_pollution.groupby(['State', 'Year'])
        .agg(
            Number_of_Highly_Polluted_Cities=('City', 'count'),
            Highly_Polluted_City_Names=('City', lambda x: ', '.join(sorted(x.unique()))),
            Average_Pollution_Index=('Pollution_Index', 'mean')
        )
        .reset_index()
        .sort_values(by='Number_of_Highly_Polluted_Cities', ascending=False)
        .reset_index(drop=True)
    )
    state_yearly_pollution.insert(0, 'Rank', state_yearly_pollution.index + 1)

    state_yearly_pollution.rename(columns={
        'State': 'State',
        'Year': 'Year',
        'Number_of_Highly_Polluted_Cities': 'Number of Highly Polluted Cities',
        'Highly_Polluted_City_Names': 'Highly Polluted City Names',
        'Average_Pollution_Index': 'Average Pollution Index'
    }, inplace=True)

    return state_yearly_pollution
```

```python
n = 10

most_polluted_state_df = rank_states_by_yearly_pollution(data).head(n)
most_polluted_state_df
```

```python
fig = px.bar(
    most_polluted_state_df,
    x='State',
    y='Average Pollution Index',
    color='Number of Highly Polluted Cities',
    facet_col='Year',
    text='Highly Polluted City Names',
    title='Year-wise Pollution Ranking by State'
)

fig.update_xaxes(tickangle=45)
fig.show()
```
